```python
# Linear Regression with Diabetes Dataset
from sklearn.datasets import load_diabetes
import pandas as pd
from sklearn.linear_model import LinearRegression

diabetes = load_diabetes()
df = pd.DataFrame(diabetes.data)
df.columns = diabetes.feature_names
df['target'] = diabetes.target
df.head()
```

```python
X = df[['age', 'sex', 'bmi', 'bp']]

y = df['target']
model = LinearRegression()
model.fit(X, y)
predictions = model.predict([[0.038, 0.05, 0.061, 0.0218]])
print("Predicted value:",predictions)
```

```python
# KMeans Clustering Example
from sklearn.cluster import KMeans
import pandas as pd

# Load the dataset
df = pd.read_csv('Sneakers.csv')
df.info()
df.head()
df.drop(columns=['image_count', 'formats', 'corrupt_files'], inplace=True)
df.head()
df.rename(columns={'class' : 'brand'}, inplace=True)
df.head()
# Select the features for clustering
X = df[['avg_width', 'avg_height', 'min_width', 'min_height', 'max_width', 'max_height']]
# Initialize KMeans
kmeans = KMeans(n_clusters=3, random_state=42)
# Fit the model
kmeans.fit(X)
# Predict the clusters
clusters = kmeans.predict(X)
# Add the cluster labels to the DataFrame
df['cluster'] = clusters
# Display the first few rows of the DataFrame with cluster labels
print(df.head())
```

```python
from sklearn.cluster import KMeans
import pandas as pd

# Load the dataset
df = pd.read_csv('Sneakers.csv')
df.head()
```

```python
df.drop(columns=['image_count', 'formats', 'corrupt_files'], inplace=True)
df.rename(columns={'class' : 'brand'}, inplace=True)
df.head()
```

```python
import matplotlib.pyplot as plt

x_axis = df['avg_width']
y_axis = df['avg_height']
color_labels, uniques = pd.factorize(df['brand'])
scatter = plt.scatter(x_axis, y_axis, c=color_labels, cmap='viridis', edgecolor='k')
plt.xlabel('Average Width')
plt.ylabel('Average Height')
plt.title('Scatter Plot of Average Width vs Average Height')
plt.show()
```

```python
X = df[['avg_width', 'avg_height', 'min_width', 'min_height', 'max_width', 'max_height']]
y = df['brand']

model = KMeans(n_clusters=3, random_state=42)
model.fit(X)
predictions = model.predict([[130, 131, 80, 75, 160, 140]])
print("Cluster centers:", model.cluster_centers_)
```
